**TASK 3-CENTRAL LIMIT THEOREM**
Perform the following steps:
• Take multiple samples of size 30
• Calculate the mean of each sample
• Store sample means
• Plot the distribution of sample means
• Observe and explain:
o How the distribution behaves
o Relation to CLT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Customer Shopping Trends.csv')
df
```

```python
#Select numeric column
data = df["Purchase Amount (USD)"]
```

```python
#Parameters for CLT
sample_size = 30
num_samples = 1000
```

```python
#Generate sample means
sample_means = []
for _ in range(num_samples):
    sample = data.sample(n=sample_size, replace=True)
    sample_means.append(sample.mean())
```

```python
#Plot distribution of sample means
plt.figure()
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Display basic statistics
np.mean(sample_means), np.std(sample_means)
```

**Related to CLT**
The Central Limit Theorem is verified by taking multiple samples of size 30 and plotting their means. The resulting distribution of sample means is approximately normal, centered around the population mean. This demonstrates that sample means are reliable estimators of the population mean, even when the original data is not normally distributed.

**Final Conclusion**
Sample means follow a normal distribution

Mean of sample means ≈ population mean

Variability reduces as sample size increases

CLT allows statistical inference using samples

**TASK 5-Z-SCORE CALCULATION**

Perform the following:
• Calculate Z-Score for each value in the selected column
• Add a new column named Z_Score
• Identify:
o Values with Z > 3 or Z < -3
• Mark them as outliers

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Customer Shopping Trends.csv')
df
```

```python
#Select numeric column
column = "Purchase Amount (USD)"
```

```python
#Calculate mean & standard deviation
mean_val = df[column].mean()
std_val = df[column].std()
```

```python
#Calculate Z-score
df["Z_Score"] = (df[column] - mean_val) / std_val
```

```python
#Display first few rows
df.head()
```

```python
#Identify outliers
outliers = df[(df["Z_Score"] > 3) | (df["Z_Score"] < -3)]
outliers
```

**FINAL CONCLUSION**
Z-score measures how far a value is from the mean in terms of standard deviations. Values with Z-score greater than +3 or less than −3 are considered outliers. In this dataset, no values exceed this threshold, indicating that the data does not have extreme outliers and is well distributed.

**TASK 6-BUSINESS INSIGHTS**


**Why is sampling required in real-world data analysis?**
Real-world datasets are often very large (millions of records).

Analyzing the full population takes more time, cost, and computing power.

Sampling helps us analyze a smaller portion and still get reliable insights.

It allows quick decision-making without waiting for full data processing.
**How does CLT help in analytics?**
**CLT (Central Limit Theorem) helps:**

CLT says that when we take many samples of sufficient size (usually ≥30),
the distribution of sample means becomes normal, even if the original data is not.

This makes it easier to:

Estimate population mean

Apply statistical tests

Make accurate predictions

 Business benefit: Reliable conclusions using small samples instead of huge datasets.

**Why is normal distribution important before hypothesis testing?**
Many statistical tests (like t-test, z-test, ANOVA) assume the data follows a normal distribution.
Normal distribution:
Is symmetrical
Has most values around the mean
Makes probability calculations accurate
If data is not normal, test results may be misleading or incorrect.
Business benefit: Ensures decisions (approval, rejection, forecasting) are statistically valid.

**How does Z-Score help in identifying unusual values?**
Z-Score tells how far a value is from the mean, measured in standard deviations.
Formula:

Z=(Value−Mean)/Standard Deviation
	​
Rules:
Z > +3 or Z < −3 → Unusual / Outlier
Helps identify:
Data entry errors
Fraudulent transactions
Abnormally high/low sales or scores

Business benefit: Detects risks, anomalies, and unusual behavior early.
